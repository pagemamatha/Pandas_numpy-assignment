```python
import pandas as pd

# Example dataset
data = {
    'product': ['Shirt', 'Shoes', 'Watch', 'Bag', 'Jacket'],
    'actual_price': [1200, 2500, 3000, 1800, 4000],
    'discounted_price': [900, 2000, 2700, 1500, 3500]
}

df = pd.DataFrame(data)

# Calculate average actual and discounted prices
avg_prices = df[['actual_price', 'discounted_price']].mean()

print(avg_prices)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
df = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Home', 'Beauty', 'Toys',
                 'Electronics', 'Clothing', 'Home', 'Beauty', 'Toys',
                 'Electronics', 'Clothing', 'Home', 'Beauty', 'Sports'],
    'Discount Percentage': [20, 50, 35, 40, 25, 60, 70, 55, 45, 30, 65, 80, 50, 60, 10]
})

#  Identify top 5 largest categories by product count
top_5_categories = df['Category'].value_counts().head(5).index

#  Filter only those categories
filtered_df = df[df['Category'].isin(top_5_categories)]

#  Create box plots
plt.figure(figsize=(6, 6))
filtered_df.boxplot(column='Discount Percentage', by='Category', grid=False)

#  Add titles and labels
plt.title('Distribution of Discount Percentage Across Top 5 Categories')
plt.suptitle('')  # Remove automatic Pandas title
plt.xlabel('Category')
plt.ylabel('Discount Percentage')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset
df = pd.DataFrame({
    'Category': ['Electronics', 'Clothing', 'Home', 'Beauty', 'Toys',
                 'Electronics', 'Clothing', 'Home', 'Beauty', 'Toys',
                 'Electronics', 'Clothing', 'Home', 'Beauty', 'Sports'],
    'Discount Percentage': [20, 50, 35, 40, 25, 60, 70, 55, 45, 30, 65, 80, 50, 60, 10]
})

#  Identify top 5 largest categories by frequency
top_5_categories = df['Category'].value_counts().nlargest(5).index

#  Create a list of discount lists for each top category
data_to_plot = [df.loc[df['Category'] == cat, 'Discount Percentage'] for cat in top_5_categories]

#  Create the boxplot manually
plt.figure(figsize=(9, 6))
plt.boxplot(data_to_plot, patch_artist=True,
            boxprops=dict(facecolor='orange', color='blue'),
            medianprops=dict(color='red', linewidth=1.5))

#  Customize the axes
plt.xticks(range(1, len(top_5_categories) + 1), top_5_categories, rotation=15)
plt.title('Discount Percentage Distribution Across Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Discount Percentage')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
#  Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Actual Price': [100, 200, 150, 300],
    'Discounted Price': [80, 120, 100, 150],
    'Listed Discount (%)': [20, 40, 33, 60]
})

#  Calculate the actual discount percentage using NumPy
# Formula: ((Actual - Discounted) / Actual) * 100
calculated_discount = ((df['Actual Price'] - df['Discounted Price']) / df['Actual Price']) * 100

#  Compare with the listed discount (allow small rounding differences)
mismatch = np.abs(calculated_discount - df['Listed Discount (%)']) > 1  # tolerance = 1%

#  Count mismatched records
num_mismatched = np.sum(mismatch)

print("Number of products with inconsistent discount values:", num_mismatched)
```

```python

import numpy as np

# Example data
ratings = np.array([1, 2, 3, 4, 5])
rating_counts = np.array([2, 3, 10, 25, 60])  # rating percentages or counts

# Manual weighted average calculation
weighted_sum = np.sum(ratings * rating_counts)
total_weight = np.sum(rating_counts)
weighted_average = weighted_sum / total_weight

print(f"Overall Weighted Average Rating: {weighted_average:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset of product ratings
df = pd.DataFrame({
    'Product': ['A','B','C','D','E','F','G','H','I','J'],
    'Rating': [5, 4, 3, 5, 2, 4, 5, 1, 3, 4]
})

# Count frequency of each rating (15)
rating_distribution = df['Rating'].value_counts().sort_index()

# Display distribution
print("Rating distribution:\n", rating_distribution)

# Plot as a bar chart
plt.figure(figsize=(6, 4))
plt.bar(rating_distribution.index, rating_distribution.values, color='skyblue', edgecolor='black')
plt.xlabel('Rating Value')
plt.ylabel('Number of Products')
plt.title('Distribution of Ratings')
plt.xticks([1, 2, 3, 4, 5])
plt.show()

```

```python
import pandas as pd

# Sample dataset
df_reviews = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'User': ['Alice', 'Bob', 'Charlie', 'David'],
    'Review': [
        'Good product',
        'Excellent quality, very satisfied',
        'Not bad',
        'I absolutely love this product! Highly recommend it to everyone.'
    ]
})

# Calculate the length of each review
review_lengths = df_reviews['Review'].str.len()

# Find the index of the longest review
longest_review_idx = review_lengths.idxmax()

# Retrieve the product, user, and review
longest_review = df_reviews.loc[longest_review_idx]

print("Product with longest review:", longest_review['Product'])
print("User who wrote it:", longest_review['User'])
print("Review content:", longest_review['Review'])
```

```python
import pandas as pd

# Sample dataset
df = pd.DataFrame({
    'Product': ['A','B','C','D','E','F','G','H','I','J'],
    'Rating': [5,4,3,5,2,4,5,1,3,4],
    'Rating_Count': [100, 50, 20, 200, 10, 70, 150, 5, 30, 60]
})

# Determine the 90th percentile (top 10%) and 10th percentile (bottom 10%)
top_10_percentile = df['Rating_Count'].quantile(0.9)
bottom_10_percentile = df['Rating_Count'].quantile(0.1)

# Filter high rating count products (top 10%)
high_rating_products = df[df['Rating_Count'] >= top_10_percentile]

# Filter low rating count products (bottom 10%)
low_rating_products = df[df['Rating_Count'] <= bottom_10_percentile]

# Calculate average rating for each group
avg_high = high_rating_products['Rating'].mean()
avg_low = low_rating_products['Rating'].mean()

print(f"Average rating for top 10% high rating count products: {avg_high:.2f}")
print(f"Average rating for bottom 10% low rating count products:{avg_low:.2f}")
```

```python

import pandas as pd

# Example dataframe
data = {
    'product': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I'],
    'discounted_price': [120, 250, 300, 150, 500, 450, 200, 700, 1000],
    'rating': [4.2, 4.5, 3.9, 4.1, 4.8, 4.0, 3.5, 4.6, 4.9]
}

df = pd.DataFrame(data)

# Create 3 equal-sized bins for discounted_price
df['price_bin'] = pd.qcut(df['discounted_price'], q=3, labels=['Low', 'Medium', 'High'])

# Calculate average rating for each bin
avg_rating_by_bin = df.groupby('price_bin')['rating'].mean()

print(df)
print("\nAverage rating by price bin:")
print(avg_rating_by_bin)
```
